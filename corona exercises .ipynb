```python
corona_df['date']= pd.to_datetime(corona_df['date'], format="%Y/%m/%d")
corona_df['week_number'] = corona_df['date'].dt.week
corona_df['year'] = corona_df['date'].dt.year
```

```python
df = corona_df.groupby(['region','week_number','year']).sum()
```

```python
df1 = df.reset_index()
```

```python
df1['population'] = df1['region'].map(population)
df1['contagious_rate'] = (df1['total']/df1['population'])*100000
```

```python
for var in index:
    dictregion[var]=df1[df1['region']==var]
```
